```


from google.colab import files
uploaded = files.upload();

import pandas as pd
import datetime
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#import numpy.plot


covid_data=pd.read_csv('covid_19_india.csv')
#covid_data['Date'] = pd.to_datetime(covid_data['Date'],format='%d/%m/%Y')


#covid_data['Date']= pd.to_datetime(covid_data['Date']) 
covid_data['Date'] = pd.to_datetime(covid_data.Date,dayfirst=True) #Convert to DD/MM/YYYY to YYYY/MM/DD
covid_data['Date'] = covid_data['Date'].dt.strftime('%Y/%m/%d')
data_by_dates = covid_data.groupby("Date")[["Deaths","Confirmed","Cured"]].sum()

#data_by_dates.tail(5)
#data_by_dates.tail()
data_by_dates.head()
#data_by_dates.describe()
##data_by_dates.plot(kind="line",figsize=(20,6))
```

```
# Timeline confirmed cases in India from 30 Jan 2020 to 11 April 2020
import plotly.express as px
fig=px.bar(covid_data,x="Date",y="Confirmed",barmode='group',height=400)
fig.update_layout(title_text='Coronavirus Confirmed Cases in India on daily basis',plot_bgcolor='rgb(230,230,230)')
fig.show()
```

```
import plotly.graph_objects as go
fig=go.Figure()
fig.add_trace(go.Scatter(x=covid_data['Date'],y=covid_data['Confirmed'],mode='lines+markers',name='Confirmed'))
fig.update_layout(title_text='Coronavirus Confirmed Cases in India on daily basis',plot_bgcolor='rgb(230,230,230) ')
fig.show()
```

```
import plotly.graph_objects as go
fig=go.Figure()
fig.add_trace(go.Scatter(x=covid_data['Date'],y=covid_data['Deaths'],mode='lines+markers',name='Confirmed'))
fig.update_layout(title_text='Coronavirus Death Cases in India on daily basis',plot_bgcolor='rgb(230,230,230) ')
fig.show()
```

```
import plotly.express as px
fig=px.bar(covid_data,x="Date",y="Confirmed",color='Confirmed',orientation='v',height=600,title='Confirmed Cases in India',color_discrete_sequence=px.colors.cyclical.IceFire)
fig.update_layout(plot_bgcolor='rgb(230,230,230)')
fig.show()
```
